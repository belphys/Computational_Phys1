```python
def emiss_and_deviation(years, ecuador, colombia, peru, chile):
    """
    This function calculates the average CO2 emissions per capita
    for Ecuador, Colombia, Peru, and Chile for each year. 
    It also computes the standard deviation of the emissions.

    The results are stored in a dataframe that includes the year,
    the emissions of each country, the average emission, and the
    standard deviation.
    """
    average = []
    desviations = []
    for i in range(len(years)):
        emisiones = [ecuador[i], colombia[i], peru[i], chile[i]]
        promedio = sum(emisiones) / len(emisiones)
        varianza = sum((x - promedio) ** 2 for x in emisiones) / len(emisiones)
        desvest = varianza ** 0.5
        average.append(promedio)
        desviations.append(desvest)
    
    df = pd.DataFrame({
        'Year': years,
        'Ecuador': ecuador,
        'Colombia': colombia,
        'Peru': peru,
        'Chile': chile,
        'Average': average,
        'Stand_Deviation': desviations
    })
    print(df.head(10))
    print("\nShape:", df.shape)
    
    return df

print("Common data:")
years, ecuador, colombia, peru, chile = data_in_common_rage()

print("\nAverage annual emission CO2, and the standard deviation of the emissions")
df_results = emiss_and_deviation(years, ecuador, colombia, peru, chile)
```

#### (g) Using the average CO2 emission time series and the associated standard deviations as uncertainties, perform a linear regression using a X2 minimisation method. Write a Python function that computes the best-fit parameters of the model. Plot the resulting best-fit line together with the data and error bars

```python
def model1_linear(x, a, b):
    return a* x + b 

def model2_chi2(parameters, x, y, yerr):
    a, b = parameters
    y_model = model1_linear (x, a, b)
    chi2 = np.sum(((y - y_model)/yerr) ** 2)
    return chi2

def chi2_linear_regression(df_results):
    """
    This function performs a linear regression of the average
    CO2 emissions versus time using a chi-square minimization.
    The standard deviations are used as uncertainties.

    It computes the best-fit parameters and plots the data
    with error bars.
    """
    years = np.array(df_results["Year"])
    avg = np.array(df_results["Average"])
    stdev = np.array(df_results["Stand_Deviation"])
    #Cambio en Y/Cambio en X
    a0 = (avg[-1] - avg[0]) / (years[-1] - years[0])
    b0 = avg[0] - a0 * years[0] #Intercepto inicial en la recta
    result = opt.minimize(model2_chi2, [a0, b0], args=(years, avg, stdev))
    a_best, b_best = result.x
    chi2_red = result.fun / (len(years) - 2)
    plt.figure(figsize=(12, 6))
    plt.errorbar(years, avg, yerr=stdev, fmt='.', capsize=3, label='Average')
    x_fit = np.linspace(years[0], years[-1], 100)
    plt.plot(x_fit, model1_linear(x_fit, a_best, b_best), 'r-', label=f'Regresión: y = {a_best:.4f}x + {b_best:.4f}')
    plt.xlabel('Año')
    plt.ylabel('Emisiones promedio CO₂ (ton/cápita)')
    plt.title('Regresión lineal por minimización de χ²')
    plt.legend()
    plt.show()
    return a_best, b_best, chi2_red

a_best, b_best, chi2_red = chi2_linear_regression(df_results)
```

```python
def eval_linear_model(years, avg, stdev, a_best, b_best):
    """
    This function evaluates the goodness of fit of the linear model
    by computing the global chi-square, the degrees of freedom,
    and the reduced chi-square.
    """
    y_model = a_best * years + b_best
    chi2_total = np.sum(((avg - y_model) / stdev) ** 2)
    n_points = len(years)
    n_param = 2  # a y b
    deg_of_f = n_points - n_param
    chi2_reduced = chi2_total /deg_of_f
    
    print("Global chi2:", chi2_total)
    print("Degrees of freedom:", deg_of_f)
    print("Reduced chi2:", chi2_reduced) 

    if chi2_reduced < 0.8:
        print("The errors are overestimated")
    elif chi2_reduced < 1.2:
        print("A good fit, model can describe data")
    elif chi2_reduced < 2:
         print("An acceptable model, but not totally a good fit")
    elif chi2_reduced < 3:
        print("Not a good model, doest not describe the data well")
    else:
        print("Inadequate model")
    return chi2_total, deg_of_f, chi2_reduced

chi2_total, deg_of_f, chi2_reduced = eval_linear_model(years, avg, stdev, a_best, b_best)
```

#### (f) Propose and fit an appropriate periodic function to the hourly average sea level data to model the tidal oscillation. Implement the fit using a Python function. Based on the fitted function, determine the period of the tide.

```python
from scipy.optimize import curve_fit
#Using a sinusoidal model
def sinus_function(t, A, omega, phi, C):
    return A * np.sin(omega*t + phi) + C

def fit_sinus_function(hours, avrg_sea):

    """
    This function fits a sinusoidal model to the hourly
    average sea level data in order to model the tidal
    oscillation and estimate the tidal period.
    """
    A0 = (max(avrg_sea) - min(avrg_sea)) / 2
    omega0 = 2*np.pi/12   # mareas suelen ~12 h
    phi0 = 0
    C0 = np.mean(avrg_sea)
    p0 = [A0, omega0, phi0, C0]
    params, _ = curve_fit(sinus_function, hours, avrg_sea, p0=p0)

    A, omega, phi, C = params

    period = 2*np.pi / omega

    print("Amplitude:", A)
    print("Angular frequency:", omega)
    print("Phase:", phi)
    print("Mean sea level:", C)
    print("Estimated tidal period:", period, "hours")

    t_fit = np.linspace(0, 23, 200) #24 hours, points
    y_fit = sinus_function(t_fit, A, omega, phi, C)

    plt.figure(figsize=(10,5))
    plt.scatter(hours, avrg_sea, label="Data")
    plt.plot(t_fit, y_fit, label="Fitted model", color="red")
    plt.xlabel("Hour of the day")
    plt.ylabel("Average Sea Level")
    plt.title("Tidal Oscillation Model")
    plt.legend()
    plt.show()

    return params, period
params, period = fit_sinus_function(hours, avrg_sea)
```

The fitted period of ~12 hours confirms a semi diurnal tidal regime 
(two high tides and two low tides per day). 
This matches the theoretical M2 lunar tidal constituent (12.42 h), with the small difference due to other tidal components and local effects.

#### (g) Compute the residuals of the model (difference between the data and the fitted function). Perform a Kolmogorov–Smirnov (KS) test to evaluate whether the residuals are consistent with a normal distribution.

```python
from scipy import stats

def periodic_model(t, A, w, phi, C):
    return A*np.sin(w*t + phi) + C

def  ks_test_res(hours, avrg_sea, params):
    """
    Computes the residuals of the fitted model and performs
    a Kolmogorov-Smirnov test to evaluate if the residuals
    are consistent with a normal distribution.
    """

    fit = periodic_model(hours, *params)
    residuals = avrg_sea - fit
    residuals_std = (residuals - np.mean(residuals)) / np.std(residuals)
    ks_stat, p_value = stats.kstest(residuals_std, 'norm')

    print("KS statistic:", ks_stat)
    print("p-value:", p_value)

    return residuals, ks_stat, p_value
residuals, ks_stat, p_value = ks_test_res(hours, avrg_sea, params)
```

With a p-value of 0.57 (> 0.05), we can not reject the null hypothesis. This means that the residuals are consistent with a normal distribution.

#### (i) Construct the 2D X2 surface in the vicinity of the best-fit values for two of the model parameters (for example the amplitude and phase of the tidal oscillation). Create a Python function that evaluates X2 over a grid of parameter values around the best-fit solution.

```python
def chi2_around_bestfit(hours, avrg_sea, params):

    """
    This function evaluates the chi-square surface around
    the best-fit parameters for amplitude and phase.
    """
    A_best, w_best, phi_best, C_best = params
    A_vals = np.linspace(A_best*0.8, A_best*1.2, 50)
    phi_vals = np.linspace(phi_best-1, phi_best+1, 50)
    chi2 = np.zeros((len(A_vals), len(phi_vals)))
    
    for i, A in enumerate(A_vals):
        for j, phi in enumerate(phi_vals):
            model = periodic_model(hours, A, w_best, phi, C_best)
            chi2[i,j] = np.sum((avrg_sea - model)**2)
    return A_vals, phi_vals, chi2

A_vals, phi_vals, chi2 = chi2_around_bestfit(hours, avrg_sea, params)
```

(j) Make a high-quality, labeled figure showing the resulting 2D X2
 surface: first as a 3D surface, and then as a 2D contour map. Mark the location of the minimum and overlay contours corresponding to at least three confidence levels.

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_chi2_surface(A_vals, phi_vals, chi2, params):
    """
    This function plots the chi-square surface as a 3D surface
    and as a 2D contour map around the best-fit solution.
    """

    A_best, w_best, phi_best, C_best = params

    PHI, AMP = np.meshgrid(phi_vals, A_vals)

    chi2_min = np.min(chi2)
    levels = [
        chi2_min + 2.30,
        chi2_min + 6.17,
        chi2_min + 11.8
    ]

    # The 3D surface 
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(PHI, AMP, chi2, cmap='viridis')
    ax.scatter(phi_best, A_best, chi2_min, color='red')
    ax.set_xlabel("Phase (phi)")
    ax.set_ylabel("Amplitude (A)")
    ax.set_zlabel("Chi-square")

    ax.set_title("Chi-square Surface Around Best Fit")

    plt.show()

    # 2D contour 
    plt.figure(figsize=(8,6))

    contour = plt.contour(PHI, AMP, chi2, levels=30)

    plt.contour(PHI, AMP, chi2, levels=levels, colors=['red','orange','yellow'], linewidths=2)

    plt.scatter(phi_best, A_best, color='black', label="Minimum")

    plt.xlabel("Phase (phi)")
    plt.ylabel("Amplitude (A)")
    plt.title("Chi-square Contour Map")

    plt.legend()

    plt.colorbar(contour, label="Chi-square")

    plt.show()

plot_chi2_surface(A_vals, phi_vals, chi2, params)
```
